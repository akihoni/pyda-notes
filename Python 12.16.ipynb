# Python 的数据结构和序列

## 元组 tuple()

固定长度、不可改变的 python 序列对象。

存放在圆括号()内，使用方括号[]访问元组中的元素。

可以使用加号运算将元组串联起来。

## 列表 list[]

长度可变、内容可修改。

使用 `append` 在列表末尾添加元素，使用 `insert` 在指定位置插入元素。

`list.insert(n, 'red')` 表示在列表 list 的第 n-1 和第 n 个元素之间插入元素 red（即把元素 red 放在第 n 个位置）。

使用 `pop` 移除并返回指定位置的元素。

使用 `remove` 移除第一个指定值。

### 切片 `start:stop`

切片时包括起始位置元素，不包括结束位置元素。因此结果中包含的元素个数为 `stop - start`。

负数切片时同理，不包括结束位置元素。

比如 `seq[-6:-2]` 表示倒数第6个元素到倒数第3个元素；

`seq[-2:]` 相当于 `seq[-2:0]` 表示倒数第2个元素到倒数第1个元素。
